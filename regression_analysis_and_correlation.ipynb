```python
#lets see what happens to ice cream sales as temperature increases

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#load the data
df = pd.read_csv(r'C:\Documents\portfolio_projects\exploratory_data_analysis\eda_project_1\Ice Cream Sales - temperatures.csv')
```

```python
#see the data
df.head()
```

```python
#check data types
df.dtypes
```

```python
#stats about the dataset
df_stats = df.describe()
df_stats
```

```python
#check if theres any missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#make a scatterplot
plt.scatter(x=df['Temperature'], y=df['Ice Cream Profits'])

#label it
plt.title('Temperature and Ice Cream Profits')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')
plt.show()       
```

```python
#seaborn regression plot
sns.regplot(x='Temperature', y='Ice Cream Profits', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#now lets look at the correlation
#the default method is Pearson, which measures the linear relationship between two continous variables
#the other two methods are Kendall and Spearman
df.corr()
```

```python
#Kendall correlation
df.corr(method='kendall')
```

```python
#Spearman correlation
df.corr(method='spearman')
```

```python
#we can see that there is a high positive association between the two variables

#create a heatmap for the correlation matrix

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
sns.histplot(data=df['Temperature'], kde=True, color='orange',edgecolor='red')

# Add labels and title
plt.title("Distribution of Daily Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.histplot(data=df['Ice Cream Profits'], kde=True, color='orange',edgecolor='red')

# Add labels and title
plt.title("Distribution of Ice Cream Profits")
plt.xlabel("Profit ($)")
plt.ylabel("Frequency")

plt.show()
```
