```python
#In this project, we look at the table given in the spreadsheet, and see if there's a correlation between temperature and revenue in ice cream sales.
#Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  #adjusts the configuration of the plots

#read in the data
df = pd.read_csv(r'C:\Users\User\OneDrive - University of Manitoba\Desktop\Documents\SQL_guide\Ice Cream Sales - temperatures.csv')
```

```python
#lets look at the data, make sure it was imported correctly
df.head()
```

```python
#check the data types
df.dtypes
```

```python
#check if theres any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#sort the values in ascending order
df.sort_values(by=['Temperature'], inplace=False, ascending=True)
```

```python
#create a scatterplot
plt.scatter(x=df['Temperature'], y=df['Ice Cream Profits'])

#make the labels
plt.title('Temperature and Ice Cream Profits')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')
plt.show()       
```

```python
#Regression plot
#plot the Temperature and Ice Cream Profits using seaborn
sns.regplot(x='Temperature', y='Ice Cream Profits', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#now lets look at the correlation
#the default method is Pearson, which measures the linear relationship between two continous variables
#the other two methods are Kendall and Spearman
df.corr()
```

```python
#Kendall's Tau correlation
df.corr(method='kendall')
```

```python
#Spearman Rank correlation
df.corr(method='spearman')
```

```python
#from the correlation matrices above, we can see that there is a high positive association between the two variables

#create a heatmap for the correlation matrix

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```
